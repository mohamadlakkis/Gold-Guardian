```python
import requests

from datetime import datetime, timedelta

def fetch_data_by_day(start_date, end_date):
    api_key = "goldapi-132nl4sm3n8dsww-io"  # Replace with your API key
    symbol = "XAU"
    curr = "USD"
    headers = {
        "x-access-token": api_key,
        "Content-Type": "application/json"
    }

    # Convert start_date and end_date to datetime objects
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")

    # Initialize result list
    all_data = []

    # Loop through each day in the range
    current = start
    while current <= end:
        url = f"https://www.goldapi.io/api/{symbol}/{curr}/{current.strftime('%Y-%m-%d')}"
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            data = response.json()
            all_data.append(data)
            print(f"Fetched data for {current.strftime('%Y-%m-%d')}")
        except Exception as e:
            print(f"Error fetching data for {current.strftime('%Y-%m-%d')}: {e}")
        current += timedelta(days=1)
    
    return all_data


import csv

def save_interval_data_to_csv(data, filename="gold_interval_data.csv"):
    # Save historical data to a CSV file
    with open(filename, "w", newline="") as file:
        writer = csv.writer(file)
        # Write header
        writer.writerow(["Timestamp", "Open", "Close", "High", "Low"])
        # Write rows
        for entry in data.get('data', []):  # Adjust based on API structure
            writer.writerow([
                entry.get('timestamp'),
                entry.get('open'),
                entry.get('close'),
                entry.get('high'),
                entry.get('low')
            ])


```

```python
# Example usage
data = fetch_data_by_day(start_date="2023-01-01", end_date="2023-12-31")
save_interval_data_to_csv(data)
```
